#### **2) Looking at relationships between numerical variables using pair plots and correlation plots.**

```python
#pairplots
sns.pairplot(honeypro,diag_kind="kde")
```

```python
corr_matrix = honeypro.corr()
corr_matrix['priceperlb'].sort_values(ascending=False)
```

```python
#The 2D correlation plot
correlation = honeypro.corr() 
correlation
```

```python
#Heatmap is being used for 2-D graphical representation of data
plt.figure(figsize=(14,6))
sns.heatmap(correlation, annot=True, vmin=-1, vmax=1, fmt=".2f", cmap="Spectral")
plt.show()
```

Observations:
* Total production , stocks, production value are having high strong positive correlation with number of colonies.
* Therefore, increase in total production is thereby increasing stocks along with value of production.
* Yield per colony and priceperlb do not have a high correlation with any of the features present in the dataset.


```python
#The categorical variables: state and year
print(honeypro.state.nunique())
print(honeypro.year.nunique())
```

Observation: We have honey production data for 44 US states over a span of 15 consecutive years, from 1998 to 2012.

#### **3) Overall trend of honey production over the years.**

```python
prod_per_year = honeypro.groupby('year').totalprod.mean().reset_index()
print(prod_per_year)
```

Observation: From above, we conclude that there is a huge decline in honey production over the years.

```python
#Viewing the trend over the years
plt.figure(figsize=(10,4))
sns.pointplot(x='year', y='totalprod', data=honeypro, estimator=sum, ci=None)
plt.xticks(rotation=90) # To rotate labels of x-axis
plt.show()
```

Observations: 
* The overall honey production in the US has been decresing over the years.
* The overall honey production is directly propotional to number of colonies and average yield per colony.

#### **4) Are there any patterns observed between total honey production and value of production every year. How has the value of production, which in some sense could be tied to the demand, changed every year?**

```python
#Pattern observed between total honey production and value of production every year
sns.lmplot(x='totalprod' , y='prodvalue' , data=honeypro , hue='year')
```

```python
#Pattern observed between total honey production and value of production every year
plt.figure(figsize=(5,5))
plt.scatter(x='totalprod', y='prodvalue', data=honeypro)
plt.xlabel('total honey production')
plt.ylabel('Production value')
plt.grid(False)
```

Observations:
* The variation of total honey production and value of production over the years has been inversely related to each other.
* Though total production was lowest in 2012, even then the value of production has been the highest.


```python
#Change in value of production every year
plt.figure(figsize=(10,4))
sns.pointplot(x='year', y='prodvalue' , data=honeypro , ci=None)
plt.xticks(rotation=90)
```

Observations:
* The value of production has increased from 1998 to 2012.
* The highest value of production has been recorded for the year 2012.
* It is interesting to note that since the total production has decreased over the years, the production value has been increasing over the years. 

#### **5) Variation in number of colonies over the years**

```python
sns.pointplot(x='year', y='numcol', data=honeypro, ci=None, estimator=sum)
plt.xticks(rotation=90)
```

Observations:
* The number of colonies reached its maximum in the year 2010.
* The number of colonies were lowest in the year 2008.
* As compared to the year 1998, the number of coloies have decreased in 2012.

#### **6)Variation of yield per colony over the years and the production trend at state level** 

```python
sns.pointplot(x='year', y='yieldpercol', data=honeypro, ci=None, estimator=sum)
plt.xticks(rotation=90) #variation of yield per colony over the years
```

Obsevations for yield per colony:
* Yield per colony has been the highest in the year 1998.
* Since 1998, the yield per colony has been on a decline in the following years.
* From this, we can also infer that the decline in total honey production has been caused by the decrease in yield per colony and not number of colonies.

```python
#production trend at state level
#we take "state" as hue parameter here
plt.figure(figsize = (20,10))
sns.pointplot(x='year', y='totalprod', data=honeypro, ci=None, estimator=sum ,hue='state')
plt.legend(bbox_to_anchor=(1,1))
plt.xticks(rotation=90) 
```

```python
#observing production trend at state level
sns.catplot(x='year', y='totalprod', data=honeypro , estimator=sum , col='state' , kind="point" , height=3 , col_wrap=5)
```

Observation for production trend at state level:
* ND shows the highest production of honey over all the states.
* CA shows the largest variation in honey production trend.
* Most prominent honey production states are TX, SD, ND, MT, FL, CA.
* MN shows an increase in honey production.

#### **7)Viewing the effect declining production trend has had on value of production.**

```python
sns.pointplot(x='year', y='totalprod', data=honeypro, ci=None)
plt.xticks(rotation=90)
```

```python
sns.pointplot(x='year', y='prodvalue' , data=honeypro , ci=None)
plt.xticks(rotation=90)
```

Observations:
* Although the production trend declined over the years, but the value of production has been on an increase.
* Despite overall honey production being lowest in the year 2012, the value of production marked that year has been the highest.
* Though supply declined, demand added to the increase in value of production over the years.

#### **CONCLUSION:-**

* We can conclude that the total honey production has declined over the years whereas the value of production per pound has increased.
* The reason for the declined honey production is the decrease in the yield per colony over the years.
* The highest honey producing state has been North Dakota.
* Some of the major honey producing states are California, Florida, North Dakota, South Dakota and Montana.
